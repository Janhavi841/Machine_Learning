```python
data = pd.concat([data, am_dummies, gear_dummies],axis = 1)
```

```python
data.head()
```

```python
data = data.drop(['am','gear'],axis = 1)
```

```python
data.head()
```

```python
data['mpg'].median()
```

```python
data.head()
```
